```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=300, noise=0.1, random_state=43)
```

```python
from sklearn.cluster import DBSCAN,KMeans
```

```python
df = pd.DataFrame(X, columns=['Feature_1', 'Feature_2'])
df.head()
```

```python
scalar = StandardScaler()
scaled_data = scalar.fit_transform(df)
```

```python
Kmeans = KMeans(n_clusters=2,random_state=42)
k_means_final = Kmeans.fit(scaled_data)
y_kmeans = k_means_final.predict(scaled_data)
sns.scatterplot(x=df['Feature_1'], y=df['Feature_2'], hue=y_kmeans, palette='Set1', s=100)
plt.scatter(k_means_final.cluster_centers_[:, 0], k_means_final.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')
```

IRREGULAR CLUSTERS DUE TO NOISE

DBSCAN:

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
y_dbscan = dbscan.fit_predict(scaled_data)
sns.scatterplot(x=df['Feature_1'], y=df['Feature_2'], hue=y_dbscan, palette='Set1', s=100)  
plt.title('DBSCAN Clustering Results')
```
