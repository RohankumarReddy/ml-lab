.loc  use labels

.iloc  use index numbers (positions)

| Feature             | `.loc`                              | `.iloc`                                |
| ------------------- | ----------------------------------- | -------------------------------------- |
| Meaning             | **Label-based** selection           | **Index-position based** selection     |
| Selects by          | Row **labels** and column **names** | Row **numbers** and column **numbers** |
| Includes end index? | **Yes**                             | **No** (Python-style)                  |
| Returns             | Series or DataFrame                 | Series or DataFrame                    |
| Used when           | You know labels (index names)       | You know positions (0,1,2)            |

MERGING JOINING AND CONCAT

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on="employee_id")
```

on = It is a common col taken from both the tables

```python
pd.merge(employees,salaries,on="employee_id",how="inner")
```

```python
pd.merge(employees,salaries,on="employee_id",how="outer")
```

| Join Type | Description          | Which rows are kept?                                    |
| --------- | -------------------- | ------------------------------------------------------- |
| **inner** | Intersection of keys | Only rows with **matching keys in both DataFrames**     |
| **outer** | Union of keys        | **All rows** from both DataFrames; missing values  NaN |

Agrregation= mean , median , mode, std

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

PIVOT TABLES

A pivot table is used to:

Group data

Summarize it using functions (sum, mean, count, max, etc.)

Rearrange rows and columns

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values='Sales',index='Region', columns='Product', aggfunc='median')
```

```python
pd.pivot_table(df,values=['Sales','Units'],index='Region', columns='Product', aggfunc='mean')
```
