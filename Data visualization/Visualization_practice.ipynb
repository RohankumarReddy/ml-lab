```python
q = df['match_winner'].value_counts()
sns.barplot(y = q.index, x = q.values)
```

```python
# Group by toss_decision and match_winner, then count
a = df.groupby(['toss_decision', 'match_winner']).size().reset_index(name='count')

# Plot
sns.barplot(data=a, x='count', y='match_winner', hue='toss_decision')
```

```python
qq=df['player_of_the_match'].value_counts().head(15)
sns.barplot(y = qq.index, x = qq.values)
```

```python
d = df['top_scorer'].value_counts().head(7)
sns.barplot(y = d.index, x = d.values)
```

```python
# Pick top 7 rows (or sort as needed)
d = df.groupby(['top_scorer', 'highscore']).size().reset_index(name='count')

# Plot with highscore as x-axis
sns.barplot(data=d, x='highscore', y='top_scorer')
```

```python
fig = px.bar(d, x='highscore', y='top_scorer', orientation='h', title='Top Scorers by Highscore')
fig.show()
```
