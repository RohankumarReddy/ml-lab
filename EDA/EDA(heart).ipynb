```python
# Step 1: Calculate mean of non-zero cholesterol values
mean_chol = df.loc[df['Cholesterol'] != 0, 'Cholesterol'].mean()

# Step 2: Replace zeros with this mean
df['Cholesterol'].replace(0, mean_chol, inplace=True)

# Step 3: Check result
df['Cholesterol']
```

```python
# Step 1: Calculate mean of non-zero cholesterol values
mean_chol = df.loc[df['RestingBP'] != 0, 'RestingBP'].mean()

# Step 2: Replace zeros with this mean
df['RestingBP'].replace(0, mean_chol, inplace=True)

# Step 3: Check result
df['RestingBP']
```
