Z-TEST

```python
import numpy as np
from statsmodels.stats.weightstats import ztest  
```

```python
marks = [72, 68, 75, 70, 74, 69, 71, 73, 72, 70,
         74, 76, 68, 69, 71, 73, 72, 74, 70, 71,
         73, 72, 74, 75, 69, 68, 70, 71, 72, 73]

z,p = ztest(marks,value =70)
print(f"Z Stat:{z}")
print(f"P Value:{p}")
if p < 0.05:
    print("Reject Null Hypothesis")

```

| Term                   | In your code            |
| ---------------------- | ----------------------- |
| Sample                 | `marks` list            |
| Sample size (n)        | 30                      |
| Population             | Assumed, not given      |
| Population mean (μ)    | `70`                    |
| Standard deviation (σ) | Calculated from `marks` |
| Test used              | One-sample Z-test       |


2 Sample T Test

```python
from scipy.stats import ttest_ind

class_A = [70, 72, 68, 74, 71, 69]
class_B = [65, 67, 66, 64, 68, 63]
t,p = ttest_ind(class_A, class_B)
print(f"T Stat:{t}")
print(f"P Value:{p}")
if p < 0.05:
    print("Reject Null Hypothesis")
```

| Test             | Null Hypothesis           |
| ---------------- | ------------------------- |
| Z / T (1 sample) | Mean = given value        |
| T (2 samples)    | Mean₁ = Mean₂             |
| Paired T         | Before = After            |
| Chi-square       | Variables are independent |
| ANOVA            | All means are equal       |
| Correlation      | No relationship           |

```python
import pandas as pd
from scipy.stats import chi2_contingency

data = {
    "Survived": [80, 100],
    "Died": [120, 20]
} #contingency table

df = pd.DataFrame(data, index=["Male", "Female"])
chi2, p, dof, expected = chi2_contingency(df)
print("Chi-square =", chi2)
print("p =", p)
print("Expected:\n", expected)
if p < 0.05:
    print("Reject Null Hypothesis")
```

ANOVA TEST

```python
from scipy.stats import f_oneway

group1 = [70, 72, 68, 71]
group2 = [65, 67, 66, 64]
group3 = [80, 82, 79, 81]

f, p = f_oneway(group1, group2, group3)

print("F-value:", f)
print("p-value:", p)

if p <= 0.05:
    print("Reject H0")
else:
    print("Fail to reject H0")
```
