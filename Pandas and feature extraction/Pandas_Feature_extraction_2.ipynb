```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == ")":
            check = False
            return data
        if check == True:
            data = data + i
        if i == '(':
            check = True
```

```python
df["Episodes"] = df["Title"].apply(extract_episodes)
```

```python
df['Episodes']= df['Episodes'].str.replace(" eps","")
```

```python
df['Episodes'] = df['Episodes'].astype(int)
```

Extract timestamp

```python
def extract_time(txt):
    check = False
    data = ""
    count = 0

    for i in txt:
        if i == ')':
            check = True
            continue

        if check == True:
            data = data + i
            count = count + 1
            if count == 18:
                return data
```

```python
df['Total Time'] = df['Title'].apply(extract_time)
```
