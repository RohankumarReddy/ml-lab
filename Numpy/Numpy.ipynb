Deep Copy = No change in original even if you modify the copy (subset or not).

Meaning:

You take a part of the original (subset), but create a deep copy of it

If you modify that deep copy
â†’ the original will NOT change


ARITHMATIC OPERATION === DEEP COPY




```python
shallow = arr1[1:4]
shallow[0]=999
print("Original",arr1)
print("Shallow:",shallow)
```

| **array_split** | general split    | any axis     | any dimension | no restriction      |
