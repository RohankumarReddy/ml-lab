


Suppose you collected the **heights of 50 students** in a class (in cm):

```
150, 152, 153, 155, 156, 156, 157, 158, 160, 161, 162,  
163, 164, 164, 165, 166, 167, 168, 170, 171, ... (50 values)
```

### Step 1: **Create bins**

Divide the data into height intervals, for example:

* 150155
* 155160
* 160165
* 165170
* 170175

### Step 2: **Count frequency**

Count how many students fall in each bin:

| Height Range (cm) | Frequency |
| ----------------- | --------- |
| 150155           | 5         |
| 155160           | 10        |
| 160165           | 15        |
| 165170           | 12        |
| 170175           | 8         |

These frequencies would be drawn as **bars** in the histogram portion of the distribution plot.

---

### Step 3: **Draw density curve (KDE)**

A smooth curve is drawn over the bars that shows the **overall shape** of the height distribution.

In this example:

* The KDE curve will be **single-peaked** (unimodal)
* It will look **bell-shaped**
* The peak will be around **160165 cm**, since this range has the highest frequency.

---

#  **Interpretation**

From this distribution plot, we can say:

* Most students are between **160170 cm** tall.
* The distribution is **almost normal** (bell-shaped).
* There are no extreme outliers.
* The variation is moderate, heights are not too spread out.

BOX PLOT

5 num summary:

A box plot is created from 5 key statistics:

1 Minimum (smallest value)


2 Q1 (1st Quartile = 25th percentile)


3 Median (Q2 = 50th percentile)


4 Q3 (3rd Quartile = 75th percentile)


5 Maximum (largest value

Q1 (First Quartile)  25% point

25% of the data is below Q1

Also called the lower quartile

Q2 (Second Quartile)  50% point

This is the median

50% of data is below, 50% above

Q3 (Third Quartile)  75% point

75% of the data is below Q3

Also called the upper quartile

````python
sns.boxplot(x='smoker',y='total_bill',data=df)
````
